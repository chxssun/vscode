### DataFrame실습2
- score.csv

```python
# data 읽어오기 -> score변수
# 과목 인덱스로 변경
# encoding = "euc-kr"
score = pd.read_csv("data/score.csv", index_col="과목", encoding="euc-kr")
score
# 한글 인코딩 3가지
# euc-kr, utf-8, cp949
```

```python
# 학급별 점수 총합
score.sum()
# 축설정 -> 기본 axis = 0 (행끼리 연산)
```

```python
# 과목별 합계 (열끼리 연산) -> axis = 1
score.sum(axis=1)
```

1. 과목별 합계를 구하여 DF에 추가("합계" 컬럼추가)

```python
score["합계"] = score.sum(axis=1)
score
```

2. 과목별 평균을 구해서 "평균" 컬럼 추가

```python
score["평균"] = score.iloc[:,:4].mean(axis=1)
score
```

3. 반별 평균을 계산해서 '반평균' 행 추가하기

```python
score.loc["반평균"] = score.mean()
score
```

- 의미없는 데이터 -로 변경하기

```python
score.loc["반평균","합계":] = "-"
score
```

4. 과목별로 가장 높은 점수와 가장 낮음 점수의 차이
    - max()
    - min()

```python
#1. 과목별 비교를 통해 가장 높은점수 구하기 (max_score)
#2. 과목별 가장 낮은 점수 구하기 (min_score)
#3. max_score - min_score
max_score = score.iloc[:5,:4].max(axis=1)
min_score = score.iloc[:5,:4].min(axis=1)
max_score - min_score
```

### apply 함수
- 행 or 열 단위로 함수를 적용할 때 사용한다
- pandas 객체에 다른 함수를 적용할 때 사용
- df.apply(함수명, 축설정/axis=0 or 1)

```python
# 최댓값에서 최솟값을 빼주는 함수
def max_min (x):
    return x.max() - x.min()
```

```python
# 함수 적용하기
score.iloc[:5,:4].apply(max_min, axis=1)
```

### 데이터 병합하기
- pd.concat([df1,df2...])

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3']
                   }, index = [0,1,2,3])

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                    'B':['B4','B5','B6','B7'],
                    'C':['C4','C5','C6','C7']},
                  index= [0,1,2,3])

df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                    'B':['B8','B9','B10','B11'],
                    'C':['C8','C9','C10','C11']},
                  index=[0,1,2,3])
```

```python
pd.concat([df1,df2,df3])
# 병합도 마찬가지로 axis = 0 기본값
# 행 방향으로 병합
# 컬럼명이 동일한 데이터끼리 병합
```

```python
pd.concat([df1,df2,df3], axis=1)
# axis = 1 -> 열방항으로 병합
# 인덱스값이 동일한 데이터끼리 병합
```
