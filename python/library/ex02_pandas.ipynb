```python
#   
score.sum()
#  ->  axis = 0 ( )
```

```python
#   ( ) -> axis = 1
score.sum(axis=1)
```

1.    DF ("" )

```python
score[""] = score.sum(axis=1)
score
```

2.    ""  

```python
score[""] = score.iloc[:,:4].mean(axis=1)
score
```

3.    ''  

```python
score.loc[""] = score.mean()
score
```

-   - 

```python
score.loc["","":] = "-"
score
```

4.

    - min()

```python
#1.       (max_score)
#2.      (min_score)
#3. max_score - min_score
max_score = score.iloc[:5,:4].max(axis=1)
min_score = score.iloc[:5,:4].min(axis=1)
max_score - min_score
```

- pd.concat([df1,df2...])

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3']
                   }, index = [0,1,2,3])

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                    'B':['B4','B5','B6','B7'],
                    'C':['C4','C5','C6','C7']},
                  index= [0,1,2,3])

df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                    'B':['B8','B9','B10','B11'],
                    'C':['C8','C9','C10','C11']},
                  index=[0,1,2,3])
```

```python
pd.concat([df1,df2,df3])
#   axis = 0 
#   
#    
```

```python
pd.concat([df1,df2,df3], axis=1)
# axis = 1 ->  
#    
```
