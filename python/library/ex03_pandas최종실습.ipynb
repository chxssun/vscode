```python
import pandas as pd
#   (data2019, data2020, data2021)
#  :    ,  encoding = 'euc-kr'
data2019 = pd.read_csv("data/2019.csv", encoding="euc-kr", index_col="")
data2020 = pd.read_csv("data/2020.csv", encoding="euc-kr", index_col="")
data2021 = pd.read_csv("data/2021.csv", encoding="euc-kr", index_col="")
data2019
```

```python
# 2021  "" 
data2021 = data2021.drop("")
# data2021.drop("", inplace=True) :   
data2021
```

```python
#      -> ""  
data2019[""] = data2019.iloc[:,1:].sum(axis=1)
data2020[""] = data2020.iloc[:,1:].sum(axis=1)
data2021[""] = data2021.iloc[:,1:].sum(axis=1)
data2019
```

```python
#    ""    (temp2019,temp2020,temp2021)
temp2019 = data2019[data2019[""] == ""].iloc[:,1:]
temp2020 = data2020[data2020[""] == ""].iloc[:,1:]
temp2021 = data2021[data2021[""] == ""].iloc[:,1:]
temp2019
```

```python
#     Series (s2019, s2020, s2021)
s2019 = temp2019[""]
s2020 = temp2020[""]
s2021 = temp2021[""]
s2019
```

```python
# Series  (?    ) 2019, 2020, 2021
s2019.name = "2019"
s2020.name = "2020"
s2021.name = "2021"
s2019
```

```python
#  (2019-2020, 2020-2021)
change2019_2020 = (s2020 - s2019)/s2019 *100
change2020_2021 = (s2021 - s2020)/s2020 *100
#  (2019-2020, 2020-2021)
change2019_2020.name = "2019_2020 "
change2020_2021.name = "2020_2021 "
change2019_2020
```

```python
#  
pd.concat([s2019, change2019_2020, s2020, change2020_2021, s2021], axis=1)
```
