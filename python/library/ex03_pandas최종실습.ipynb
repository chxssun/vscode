```python
import pandas as pd
# 데이터 불러오기 (data2019, data2020, data2021)
# 조건 : 관서명 컬럼을 인덱스로 설정, 인코딩 encoding = 'euc-kr'
data2019 = pd.read_csv("data/2019.csv", encoding="euc-kr", index_col="관서명")
data2020 = pd.read_csv("data/2020.csv", encoding="euc-kr", index_col="관서명")
data2021 = pd.read_csv("data/2021.csv", encoding="euc-kr", index_col="관서명")
data2019
```

```python
# 2021년도에만 존재하는 "광주지방경찰청" 삭제
data2021 = data2021.drop("광주지방경찰청")
# data2021.drop("광주지방경찰청", inplace=True) : 재할당안해도 삭제후 저장됨
data2021
```

```python
# 각 연도별 범죄현황 총합 구하기 -> "총합" 컬럼 추가
data2019["총합"] = data2019.iloc[:,1:].sum(axis=1)
data2020["총합"] = data2020.iloc[:,1:].sum(axis=1)
data2021["총합"] = data2021.iloc[:,1:].sum(axis=1)
data2019
```

```python
# 각 연도별로 구분이 "발생건수" 인 데이터만 추출해서 임시변수(temp2019,temp2020,temp2021)담기
temp2019 = data2019[data2019["구분"] == "발생건수"].iloc[:,1:]
temp2020 = data2020[data2020["구분"] == "발생건수"].iloc[:,1:]
temp2021 = data2021[data2021["구분"] == "발생건수"].iloc[:,1:]
temp2019
```

```python
# 각연도별로 총합 컬럼만 추출하기 Series형태로 추출될것(s2019, s2020, s2021)담기
s2019 = temp2019["총합"]
s2020 = temp2020["총합"]
s2021 = temp2021["총합"]
s2019
```

```python
# Series 데이터 이름변경(왜? 총합할껀데 이름이 다 총합이니까) 2019총계, 2020총계, 2021총계
s2019.name = "2019총계"
s2020.name = "2020총계"
s2021.name = "2021총계"
s2019
```

```python
# 증감률 구하기(2019-2020, 2020-2021)
change2019_2020 = (s2020 - s2019)/s2019 *100
change2020_2021 = (s2021 - s2020)/s2020 *100
# 이름 바꿔주기(2019-2020증감률, 2020-2021증감률)
change2019_2020.name = "2019_2020 증감률"
change2020_2021.name = "2020_2021 증감률"
change2019_2020
```

```python
# 데이터 병합하기
pd.concat([s2019, change2019_2020, s2020, change2020_2021, s2021], axis=1)
```
