### 들여쓰기의 중요성

```python
if True:
    print("실행문장 실행")
print("if문 바깥쪽 실행문장")
```

```python
if False:
    print("실행문장 실행")
print("if문 바깥쪽 실행문장")
```

```python
num1 = int(input("첫번째 정수 입력 >> "))
num2 = int(input("두번째 정수 입력 >> "))
if num1 > num2 :
    print("첫번째 정수가 더 큽니다")
elif num1 < num2 :
    print("두번째 정수가 더 큽니다")
else :
    print("두 정수는 같습니다")
```

```python
score = int(input("점수 입력 >> "))
grade = ""
if 100 >= score >= 90 :
    grade = "A"
elif score >= 80 :
    grade = "B"
elif score >= 70 :
    grade = "C"
elif score >= 60 :
    grade = "D"
else :
    grade = "F"
print(f"{score}점은 {grade}학점 입니다.")
```

```python
print('''
================================================
================---- 자판기 ----================
================================================
== 1.콜라 == 2.사이다 ==  3.물  =======insert===
==--------==----------==--------=======------===
==   600  ==    800   ==  1000  ======= O00원===
=======================================------===
================================================
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==+++++++++++++++      AIS     +++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
================================================
================================================
  ===                                      ===
  ===                                      ===
''')
coin = int(input("insert coin >> "))
menu = int(input("메뉴선택 >> "))
if menu == 1:
    coin -= 600
elif menu == 2:
    coin -= 800
elif menu == 3:
    coin -= 1000
else:
    print("메뉴를 잘못 입력 하셧습니다")
if coin < 0 :
    print("잔돈이 부족합니다")
else:
    print(f"잔돈:{coin} \n천원:{coin//1000}개, 오백원:{coin%1000//500}개, 백원:{coin%1000%500//100}개")
```
