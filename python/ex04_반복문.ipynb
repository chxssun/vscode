```python
# 1부터 3까지 출력
num = 1
while num <= 3:
    print(num)
    num += 1
```

```python
# break 반복문을 중단하는 기능
while True:
    print("반복할께")
    break
```

```python
num = 1
while True:
    print(num)
    num += 1
    if num > 3:
        break
```

```python
import random as rd 
ran = rd.randint(1, 50)
while True:
    num = int(input("숫자를 입력하세요 >> "))
    if ran > num:
        print(f"{num}보다 큰 수 입니다")
    elif ran < num:
        print(f"{num}보다 작은 수 입니다")
    else:
        print(f"{num} 정답입니다")
        break
```

```python
hi = "안녕하세요~오"
for i in hi:
    print(i, end="❤️")    
# print문의 end속성 -> 기본값은 개행(\n)
```

```python
score_list = [90,45,70,60,55]
num = 1
for i in score_list:
    if i >= 60:
        print(f"{num}번 학생은 합격입니다")
    else:
        print(f"{num}번 학생은 불합격입니다")
    num += 1    
```
