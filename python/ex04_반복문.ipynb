```python
import random as rd 
ran = rd.randint(1, 50)
while True:
    num = int(input("  >> "))
    if ran > num:
        print(f"{num}   ")
    elif ran < num:
        print(f"{num}   ")
    else:
        print(f"{num} ")
        break
```

```python
score_list = [90,45,70,60,55]
num = 1
for i in score_list:
    if i >= 60:
        print(f"{num}  ")
    else:
        print(f"{num}  ")
    num += 1    
```
