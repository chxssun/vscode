### 문자열 인덱싱 슬라이싱
- 문자열도 인덱스 번호를 가진다
- 인덱싱 : 인덱스 번호 하나를 가리켜서 데이터 추출
- 슬라이싱 : 범위를 지정하여 여러개의 데이터를 추출

```python
name = "My name is CS"
```

```python
# n추출하기
# 주의1 : 인덱스 번호는 0부터 시작
# 주의2 : 띄어쓰기도 인덱스 번호를 가진다
# 주의3 : 뒤에서 부터도 인덱스 번호를 가진다(음수)
print(name[3])
print(name[-2])
```

```python
# 슬라이싱 방법 : 변수명[범위]
# 범위[시작인덱스 : 끝인덱스+1]
# 주의1 : 슬라이싱에서 범위를 작성할 때 끝 인덱스 범위는 내가 가져오고 싶은 인덱스 +1
print(name[:2])
print(name[3:7])
print(name[-2:])
print(name[:])
```

### 문자열 포매팅
- % 기호 포매팅
    - 단점 : 자료형에 따라 사용하는 알파벳을 암기
- format 함수 포매팅
- f문자열 포매팅
    - 장점 : 직관적이다

```python
month = 6
day = 25
today = "오늘은 "+month+"월 "+day+"일 입니다!"
today
# 문자열 연산을 할때 문자열과 숫자는 연산이 안된다!
```

```python
# 1. % 기호포매팅 -> % 기호를 이용하여 숫자 대입
# 단점 : 데이터타입에 따른 알파벳을 암기
month = 6
day = 25
today = "오늘은 %d월 %d일 입니다!"%(month, day)
today
```

```python
# 2. format 함수
# 중괄호 순서와 포맷함수 안에 들어가는 변수순서가 일치
month = 6
day = 25
today = "오늘은 {}월 {}일 입니다!".format(month, day)
today
```

```python
# 3. f문자열 포매팅
# format 함수의 단점을 보완하여 변수명을 문자열 그 위치에 작성
# 직관적이다!
month = 6
day = 25
today = f"오늘은 {month}월 {day}일 입니다!"
today
```

```python
x = 100
y = 200
add = x + y
print(f"{x}와 {y}의 합은 {add}입니다")
```

### 연산자


### 연산자의 종류


| 연산자 카테고리 | 연산자 | 설명 |
|:----------------:|:------:|:-----|
| 산술 연산자 | `+` | 덧셈 |
| | `-` | 뺄셈 |
| | `*` | 곱셈 |
| | `/` | 실수 나눗셈 |
| | `//` | 정수 나눗셈 (몫) |
| | `%` | 나머지 |
| 지수 연산자 | `**` | 지수 (제곱) |
| 대입(복합) 연산자 | `=` | 할당 |
| | `+=` | 덧셈 후 할당 |
| | `-=` | 뺄셈 후 할당 |
| | `*=` | 곱셈 후 할당 |
| | `/=` | 나눗셈 후 할당 |
| | `//=` | 몫 나눗셈 후 할당 |
| | `%=` | 나머지 나눗셈 후 할당 |
| 관계(비교) 연산자 | `>` | 크다 |
| | `>=` | 크거나 같다 |
| | `<` | 작다 |
| | `<=` | 작거나 같다 |
| | `==` | 같다 |
| | `!=` | 같지 않다 |
| 논리 연산자 | `and` | 논리곱 (AND) |
| | `or` | 논리합 (OR) |
| | `not` | 논리부정 (NOT) |
| 멤버 연산자 | `in` | 대상 객체 내에 값이 존재하는지 확인 |
| | `not in` | 대상 객체 내에 값이 존재하지 않는지 확인 |
| 삼항 연산자 | `a if 조건식 else b` | 조건식이 참이면 a, 거짓이면 b를 반환 |

```python
num1 = int(input("정수를 입력하세요 >> "))
num2 = int(input("정수를 입력하세요 >> "))
print(f" 더하기 결과 : {num1+num2} \n 빼기 결과: {num1-num2} \n 곱하기 결과 : {num1*num2} \n 나누기결과 : {num1/num2}")
```
