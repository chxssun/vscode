```python
# transformers 안에있는 pipline 기능
from transformers import pipeline
model = pipeline(task="text-classification", model="hun3359/klue-bert-base-sentiment")
text = "플라스크 수업 재미 있어요!"
model(text)
```

```python
# 플라스크와 모델 연동하기
# 라이브러리 불러오기
from flask import Flask, render_template, request
from transformers import pipeline
import json

# 모델 불러오기
model = pipeline(task="text-classification", model="hun3359/klue-bert-base-sentiment")

# 텍스트를 입력받아 감성분석하여 예측값을 돌려주는 사용자 정의 함수
def predict_emotion(text):
    result = model(text)
    return result

# 플라스크 웹 초기화 및 라우팅
app = Flask(__name__)

@app.route("/")
def home():
    return render_template("predict.html")

# predict 결과 라우트 -> post
@app.route("/predict", methods=["POST"])
def predict():
    input_text = request.form["input_text"]
    result = predict_emotion(input_text)
    # return result
    return json.dumps(result, ensure_ascii=False)
    # 한글형태로 유지해주는 코드

# 서버실행
if __name__ == "__main__":
    app.run(port = 5000, host = "192.168.20.145")
```
